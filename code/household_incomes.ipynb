```python
import hinc
import hinc2
import thinkstats2
import thinkplot
import numpy as np
```

```python
data = hinc2.main()
```

```python
data = hinc.ReadData()
```

```python
interp = hinc2.InterpolateSample(data, log_upper=6.0)
cdf = thinkstats2.Cdf(interp)
thinkplot.Cdf(cdf)
thinkplot.Show()
```

```python
median = cdf.Percentile(50)
mean = cdf.Mean()
print([median, mean])
```
